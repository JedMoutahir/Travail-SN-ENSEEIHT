```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
r = [[1, 10], [0, -15]]
N = 10
```

```python
def V(n, s, gamma):

    if n == 0:
        return 0, [s]

    if s:
        res1, state1 = V(n - 1, s, gamma)
        res2, state2 = V(n - 1, not s, gamma)
        if r[0][0] + gamma * res1:
            res = r[0][0] + gamma * res1
            states = state1 + [s]
        else:
            res = r[0][1] + gamma * res2
            states = state2 + [s]
        return res, states
    
    else:
        res1, state1 = V(n - 1, s, gamma)
        res2, state2 = V(n - 1, not s, gamma)
        if r[1][0] + gamma * res1:
            res = r[1][0] + gamma * res1
            states = state1 + [s]
        else:
            res = r[1][1] + gamma * res2
            states = state2 + [s]
        return res, states
```

```python
V(N, False, 0.1)
```

```python
# List for gamma between 0 and 1
gamma_list = np.linspace(0, 1, 100)
V_list_False = [V(N, False, gamma)[0] for gamma in gamma_list]
V_list_True = [V(N, True, gamma)[0] for gamma in gamma_list]

plt.plot(gamma_list, V_list_False, label='False in state 1')
plt.plot(gamma_list, V_list_True, label='True in state 1')
plt.title('The reward for different gamma with N = 10 max depth')
plt.xlabel('gamma')
plt.ylabel('V')
plt.legend()
plt.show()
```

```python
gamma_list = np.linspace(0, 1, 10)
V_list_False = [V(N, False, gamma)[1] for gamma in gamma_list]
V_list_True = [V(N, True, gamma)[1] for gamma in gamma_list]

# Print the lists
for i in range(len(gamma_list)):
    print("gamma = ", gamma_list[i], "False: ", V_list_False[i], "True: ", V_list_True[i])
```
