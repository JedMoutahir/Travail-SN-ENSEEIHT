```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n, m = 20, 50
# Tableau de 0 de taille n x m
V = np.zeros((n, m))
# Tableau de False de taille n x m
C = np.zeros((n, m), dtype=bool)
```

```python
V[0, :] = 0
V[:, 0] = 0

C[0, :] = False
C[:, 0] = False
```

```python
# Display V values in a 2D plot
def plot_V():
    """ Display V values in a 2D plot : The average reward in function of the number of products in stock (y-axis) and the number of days left (x-axis)"""
    plt.imshow(V, cmap='hot', interpolation='nearest')
    plt.colorbar(orientation='horizontal')
    plt.title("The average reward in function of the number of products in stock (y-axis) \n and the number of days left (x-axis)")
    plt.show()

plot_V()
```

```python
def plot_C():
    """ Display C values in a 2D plot : False = 0 for choosing the price p2 = 1, True = 1 for choosing the price p1 = 5 """
    plt.imshow(C, cmap='hot', interpolation='nearest')
    plt.title("Price chosen for the product in function of the number of products in stock (y-axis) and the number of days left (x-axis) \n black for choosing the price p2 = 1, white for choosing the price p1 = 5")
    plt.show()

plot_C()
```

```python
def V_from_prev_T_and_prev_s(T, s):
    if T==0 or s==0:
        return 0
    return max(
        0.5 + 0.9 * V[T-1, s] + 0.1 * V[T-1, s-1],
        0.8 + 0.2 * V[T-1, s] + 0.8 * V[T-1, s-1],
    ), 0.5 + 0.9 * V[T-1, s] + 0.1 * V[T-1, s-1] > 0.8 + 0.2 * V[T-1, s] + 0.8 * V[T-1, s-1]
```

```python
for T in range(1, n):
    for s in range(1, m):
        V[T, s], C[T, s] = V_from_prev_T_and_prev_s(T, s)
```

```python
plot_V()
```

```python
plot_C()
```
